```python
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(pred):
  labels=pred.label_ids
  preds=pred.predictions.argmax(-1)
  f1=f1_score(labels, preds, average="weighted")
  acc=accuracy_score(labels, preds)
  return {"accuracy":acc, "f1": f1}
```

```python
import torch
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
!export CUDA_LAUNCH_BLOCKING=1
```

```python
from transformers import AutoModelForSequenceClassification
model_ckpt="cl-tohoku/bert-base-japanese-whole-word-masking"


num_labels=5
model=AutoModelForSequenceClassification.from_pretrained(model_ckpt, num_labels=num_labels).to(device)
```

```python
from transformers import Trainer, TrainingArguments

batch_size=64
logging_steps=len(dataset_dict["train"])//batch_size
model_name="bert-japanese-finetuned-Ukraine-tweet"
training_args=TrainingArguments(output_dir=model_name,
                                num_train_epochs=3,
                                learning_rate=2e-5,
                                per_device_train_batch_size=batch_size,
                                per_device_eval_batch_size=batch_size,
                                weight_decay=0.01,
                                evaluation_strategy="epoch",
                                disable_tqdm=False,
                                logging_steps=logging_steps,
                                push_to_hub=True,
                                log_level="error")
```

```python
from transformers import Trainer

trainer=Trainer(model=model, args=training_args, compute_metrics=compute_metrics,
                train_dataset=dataset_dict["train"], eval_dataset=dataset_dict["validation"],
                tokenizer=tokenizer)
trainer.train()
```

```python
preds_output=trainer.predict(dataset_dict["validation"])
```

```python
preds_output.metrics
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_preds, y_true, labels):
  cm=confusion_matrix(y_preds, y_true, normalize="true")
  fig, ax=plt.subplots(figsize=(7.3,7.3))
  disp=ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
  disp.plot(cmap="Blues", values_format=".2f", ax=ax, colorbar=False)
  plt.title("Normalized confusion matrix")
  plt.show()
```

## エラー分析

```python
from torch.nn.functional import cross_entropy

def forward_pass_with_label(batch):
  # 全ての入力テンソルをモデルと同じデバイスに配置する
  inputs={k:v.to(device) for k,v in batch.items() if k in tokenizer.model_input_names}

  with torch.no_grad():
    output=model(**inputs)
    pred_label=torch.argmax(output.logits, axis=-1) 
    loss=cross_entropy(output.logits, batch["label"].to(device), reduction="none")

  #データセットの他の列との互換性を保つため、出力をcpu上に配置する
  return {"loss": loss.cpu().numpy(), "predicted_label": pred_label.cpu().numpy()}
```

```python
# データセットをPyTorchのテンソルに変換
dataset_dict.set_format("torch", columns=["input_ids", "attention_mask", "label"])

# 損失の値を計算
dataset_dict["validation"]=dataset_dict["validation"].map(forward_pass_with_label, batched=True, batch_size=16)
```

```python
def label_int2str(row):
  return dataset_dict["train"].features["label"].int2str(row)
```

```python
dataset_dict.set_format("pandas")
cols=["text", "label","predicted_label","loss"]
test=dataset_dict["validation"][:][cols]
test["label"]=test["label"].apply(label_int2str)
test["predicted_label"]=(test["predicted_label"].apply(label_int2str))
```

```python
# どういうラベルが間違いやすいかみたいな感じ
test.sort_values("loss", ascending=False).head(10)
```
